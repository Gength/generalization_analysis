```python
import pandas as pd
import seaborn as sns
import json
import matplotlib.pyplot as plt

with open("output/hybrid_extensive_20260514_143816.json", "r") as f:
    data = json.load(f)

df = pd.DataFrame(data['results'])

#Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="miner", y="gen_total")
plt.title("Generalization Stability Across Multiple Runs (w=0.5)")
plt.ylabel("Gen_Total Score")
plt.xlabel("Process Miner")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import json
import matplotlib.pyplot as plt

file_path = "output/hybrid_extensive_20260514_201205.json"
with open(file_path, "r") as f:
    data = json.load(f)

df = pd.DataFrame(data['results'])

plt.figure(figsize=(12, 7))

sns.lineplot(data=df, x="w_weight", y="gen_total", hue="miner", marker="o", linewidth=2.5)

plt.title("Generalization Ablation Study: Weight Sweep Analysis", fontsize=14)
plt.xlabel("Fusion Weight (w)\n(0.0 = Pure Structural Penalty | 1.0 = Pure Generative Flexibility)", fontsize=12)
plt.ylabel("Gen_Total Score", fontsize=12)
plt.ylim(0, 1.05)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Process Miner", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
